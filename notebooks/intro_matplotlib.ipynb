Bar Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/archive/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
# group by genre and calculate the average total streams
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
# Label the chart

plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")


plt.show()
```

```python
# Clean up the visual

plt.figure(figsize=(10,6))

plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
# More visual cleaning -- remove spines(borders)

plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)


ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python

avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.Figure(figsize=(10, 6))
plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = "pink", width = 0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
# Highlight max index bar

plt.Figure(figsize=(10, 6))
colors = ['#7b92c9']*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#f030ad"


plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
# Focal point highlight
plt.figure(figsize=(12, 8))
colors = ['#7b92c9']*len(avg_artist_streams)

special_artist = "Ariana Grande"
colors = ["#f030ad" if artist == special_artist else "gray" for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

Scatter Plots

```python
spotify_df.columns
spotify_df.info()
```

```python
plt.scatter(spotify_df["Streams Last 30 Days (Millions)"], spotify_df["Total Streams (Millions)"])
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"],
    spotify_df["Total Streams (Millions)"],
    color = "#17df21",
    alpha = 0.8,
    s= 25 )

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

featured_genre = "Pop"

colors = ["#17df21" if genre == featured_genre else "#d1d5d1" for genre in spotify_df["Genre"]]

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"],
    spotify_df["Total Streams (Millions)"],
    color = colors,
    alpha = 0.8,
    s= 25 )

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")
#plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Monthly Listens or Total Streams?", fontsize = 18, pad=35)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)
                            

plt.tight_layout()
plt.show()
```
